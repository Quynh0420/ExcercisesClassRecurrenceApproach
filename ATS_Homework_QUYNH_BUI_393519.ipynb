# Task 1A

Design 4 classes ( Figure, Triangle, Rectangle, Circle) basing on the description given below. Do not use build-in libraries, when you are not asked for it. Use inheritance and method overloading mechanisms, when needed. 
Let’s assume, that we have 4 types of figures {“unknown”,”triangle”,”rectangle”,”circle”}. 

<br><br>
<b>All figures should have access to </b>:

<b>attributes:</b><br>
	id – id of given figure. Random value between {1,1000}. You can use function random.Random(), which returns float value between 0 and 1 and function math.floor()<br>
	type  - type of given figure. Values {“unknown”,”triangle”,”rectangle”,”circle”}. <br>
	color   - color of given figure. Values {“red”,”blue”,”green”,”white”}. Default = “white”	<br>
<b>methods: </b><br>
	printInfo(): prints info about figure id, type, color <br>
	setColor(colorNr): sets color of the figure, where 1=white, 2=red, 3=blue,4=green <br>
    setRndColor(): sets random color . You can use function random.Random() and math.floor() <br>
    getArea(): returns area of given figure. If figure is “unknown” returns 0 <br>
    getCircuit(): returns circuit of given figure. If figure is “unknown” returns 0 <br>

<br><br>
Three classes (Triangle, Rectangle, Circle) should be extended with the attributes and methods defined as below : <br>

<br>
<b>Triangle: </b><br>
	<b>attributes:</b><br>
        sideA – length of side A (default=1) <br>
        sideB – length of side B (default=1) <br>
        side C – length of side C (default=1) <br>
    <b>methods: </b><br>
        Check() – checks triangle condition when triangle is being created. If condition is not satisfied, then the lengths of all sides are set to 1.        Condition (a+b>=c,b+c>=a,c+a>b) <br>
<br>
Hint: You can calculate area of triangle using Heron’s Formula: <br> 
area=sqrt((p-a)*(p-b)*p-c)) where p =(a+b+c)/2. Use math.sqrt(x) <br>

<br> <br>
<b>Rectangle:</b><br>
      <b>attributes:</b><br>
		side A – length of side A (default=1) <br>
        sideB – length of side B (default=1) <br>

<b>Circle: </b><br>
      <b>attributes:</b><br>
        radius - length of the radius (default=1)<br>
      <b>methods: </b>
      <br>getDiameter() – returns the length of the diameter



```python
import math
import random

class Figure: # base class
    #Constructor of the base class. We define all the fields(attributes) for given class below. 
    #Remeber to use self keyword before each attribute.
    def __init__(self, typeA, colorA = "white"):
        self.id = math.floor(1 + (1001-1)*random.random())
        self.type = typeA
        self.color = colorA
        
    def printInfo(self):
        print("ID:", self.id, "\n"
             "Type:", self.type, "\n"
             "Color:", self.color)
        
    def setColor(self, colorNr):
        if colorNr == 1:
            self.color = "white"
        elif colorNr == 2:
            self.color = "red"
        elif colorNr == 3:
            self.color = "blue"
        elif colorNr == 4:
            self.color = "green"
        else:
            print('Please use number 1, 2, 3 or 4')
        
        print("The color of figure is now:", self.color)
        
    def setRndColor(self):
        RndNr = math.floor(1 + (5-1)*random.random())
        self.setColor(RndNr)
        
    def getArea(self):
        return 0
        
    def getCircuit(self):
        return 0
    
class Triangle(Figure): 
        
    #Constructor of the child class.
    def __init__(self, colorA = "white", sideA_ = 1, sideB_ = 1, sideC_ = 1):          
            #Here we call constructor of the base class
            Figure.__init__(self, "triangle", colorA)
            
            #We set extra field of Dog class below
            self.sideA = sideA_
            self.sideB = sideB_
            self.sideC = sideC_
            
    def printInfo(self):
        print("ID:", self.id, "\n"
             "Type:", self.type, "\n"
             "Color:", self.color, "\n"
             "3 sides:", self.sideA, self.sideB, self.sideC)
    
    def Check(self):
        a = self.sideA
        b = self.sideB
        c = self.sideC
        if a+b <= c or b+c <= a or c+a <= b:
            self.sideA = 1
            self.sideB = 1
            self.sideC = 1
            print("Condition not satisfied. All the sides are reset to 1")
            
    def getArea(self):
        p = (self.sideA + self.sideB + self.sideC)/2
        
        area = math.sqrt((p - self.sideA) * (p - self.sideB) * (p - self.sideC))
        
        return area
        
    def getCircuit(self):
        return self.sideA + self.sideB + self.sideC
    
class Rectangle(Figure): 
        
    #Constructor of the child class.
    def __init__(self, colorA = "white", sideA_ = 1, sideB_ = 1):          
            #Here we call constructor of the base class
            Figure.__init__(self, "rectangle", colorA)
            
            #We set extra field of Dog class below
            self.sideA = sideA_
            self.sideB = sideB_
            
    def printInfo(self):
        print("ID:", self.id, "\n"
             "Type:", self.type, "\n"
             "Color:", self.color, "\n"
             "2 sides:", self.sideA, self.sideB)
            
    def getArea(self):
        return self.sideA * self.sideB
        
    def getCircuit(self):
        return (self.sideA + self.sideB)*2
    
class Circle(Figure): 
        
    #Constructor of the child class.
    def __init__(self, colorA = "white", radiusA = 1):          
            #Here we call constructor of the base class
            Figure.__init__(self, "circle", colorA)
            
            #We set extra field of Dog class below
            self.radius = radiusA 
            
    def printInfo(self):
        print("ID:", self.id, "\n"
             "Type:", self.type, "\n"
             "Color:", self.color, "\n"
             "Radius:", self.radius)
        
    def getDiameter(self):
        return self.radius*2
            
    def getArea(self):
        return (self.radius**2) * 3.14 
        
    def getCircuit(self):
        return 2*self.radius*3.14
```

# Task3

Please write code in Python, which verifies if the input array/list is symmetrical. Input parameters: epsilon, array.
<b>Use recursive method approach.</b> <br><br>

Remark: Two numbers are identical if the distance between them is smaller than epsilon value. 
Below, example of two different symmetrical arrays/list, epsilon=0.5:

<br>
Table1: [33.5	22.2	5	22.5	33.6]
<br>

abs(33.5 - 33,6)<epsilon

abs(22.2 - 22.5) < epsilon  

abs(5-5)<epsilon         

Table2:  [1.5	12.2	5	5	12.5	1.6]


```python
import numpy as np

def isSymmetric (listA, eps = 0.5):
    
    if isinstance(listA, (list, tuple, np.ndarray)) == False or len(listA) == 0:
        print("It is symmetrical")
        return
    
    if abs(listA[0] - listA[-1]) < eps:
        isSymmetric(listA[1:(len(listA) -1)])
    else:
        print("It is NOT symmetrical")
        return

list1 = [33.5, 22.2, 5, 22.5, 33.6] 
list2 = np.asarray([1.5, 12.2, 5, 5, 12.5, 1.6])
list3 = np.asarray([1.5, 14, 5, 5, 12.5, 1.6])
list4 = [33.5, 22.2, 5, 24, 33.6] 
isSymmetric(list1)
isSymmetric(list2)
isSymmetric(list3)
isSymmetric(list4)

# print(list1[1:(len(list1)-1)])
```

# Task 4

Please write two separate functions in Python, which calculate the expected return and standard deviation of portfolio which contains 5 assets. <br>
Your input data:
<br>
expReturnList - list of assets expected returns <br>
expStdList - list of assets std <br>
weightsLists - list of assets weights in portfolio <br> 
covMatrix - covariance matrix (list of lists),example ->covMatrix[0][1] - returns covariance of 1st and 2nd asset <br>



```python
def ER_Portfolio (ERList, weightList):
    
    myReturn = 0
    for i in range(len(ERList)):
        myReturn = myReturn + weightList[i] * ERList[i]
    
    return myReturn

ER = [0.3, 0.15, -0.2, -0.05, 0.1]
weight = [0.25, 0.4, 0.15, 0.1, 0.1]

ER_Portfolio(ER, weight)
```

```python
def Std_portfolio (StdList, weightList, covMatrix):
    
    myStd = 0
    for i in range(len(StdList)):
        myStd = myStd + (weightList[i]**2) * (StdList[i]**2)
        
    for i in range(len(covMatrix)):
        for j in range(len(covMatrix)):
            
            if i == j:
                continue
                
            myStd = myStd + weightList[i] * weightList[j] * covMatrix[i][j]
                
    return myStd

std = [0.05, 0.08, 0.07, 0.13, 0.12]
weight = [0.25, 0.4, 0.15, 0.1, 0.1]
covMatrix = [[0.05, 0.025, 0.03, 0.05, 0.04],
            [0.025, 0.08, 0.015, 0.1, 0.035],
            [0.03, 0.015, 0.07, 0.045, 0.2],
            [0.05, 0.1, 0.045, 0.13, 0.06],
            [0.04, 0.035, 0.2, 0.06, 0.12]]
#print(covMatrix)

Std_portfolio(std, weight, covMatrix)
```
